```python
bikegwang = pd.read_csv("/content/drive/MyDrive/Python/광진구 대여소 번호.csv")

bike_using1 = bikestop01[bikestop01["대여소번호"].isin(bikegwang["대여소 번호"])]
bike_using2 = bikestop02[bikestop02["대여소번호"].isin(bikegwang["대여소 번호"])]
bike_using3 = bikestop03[bikestop03["대여소번호"].isin(bikegwang["대여소 번호"])]
bike_using4 = bikestop04[bikestop04["대여소번호"].isin(bikegwang["대여소 번호"])]
bike_using5 = bikestop05[bikestop05["대여소번호"].isin(bikegwang["대여소 번호"])]
bike_using6 = bikestop06[bikestop06["대여소번호"].isin(bikegwang["대여소 번호"])]
```

```python
bikestop02["이용건수"].unique()

# \\N 값을 null 값으로 바꾸기
bikestop02["이용건수"].replace('\\N', pd.np.nan, inplace=True)
# null 값을 가진 행 제거
bikestop02["이용건수"].dropna(inplace=True)
```

```python
bike_using1 = bike_using1.groupby('대여소번호')['이용건수'].sum().reset_index()
bike_using2 = bike_using2.groupby('대여소번호')['이용건수'].sum().reset_index()
bike_using3 = bike_using3.groupby('대여소번호')['이용건수'].sum().reset_index()
bike_using4 = bike_using4.groupby('대여소번호')['이용건수'].sum().reset_index()
bike_using5 = bike_using5.groupby('대여소번호')['이용건수'].sum().reset_index()
bike_using6 = bike_using6.groupby('대여소번호')['이용건수'].sum().reset_index()
```

```python
# 6개의 데이터프레임을 리스트로 만듦
df_list = [bike_using1, bike_using2, bike_using3, bike_using4, bike_using5, bike_using6]

# concat 함수로 데이터프레임 합치기
df_concat = pd.concat(df_list, axis=0)

# groupby와 sum 함수로 대여소번호를 기준으로 이용건수 합치기
result = df_concat.groupby(['대여소번호']).sum()

result = result.sort_values(by='이용건수', ascending=False)

result.head(20)
```
